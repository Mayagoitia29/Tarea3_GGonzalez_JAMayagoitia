```python
# Código de solución estudiante 2
import numpy as np
def ecuacion_recursiva(d1:'primer valor', d2:'segundo valor',
                       N:'numero de terminos'):
    
    n = np.arange(start=3, stop=N)
    return (n-1)*d1 + (n-1)*d2

d1, d2, N = 0, 1, 10

ecu1 = ecuacion_recursiva(d1, d2, N)
ecu1
```

### Ejercicio 3:
>3. Remove all of the vowels in a string [make a list of the non-vowels].

```python
# Código de solución estudiante 1
```

```python
#Código de solución de estudiante 2
def non_vowels(variable):
    return [i for i in variable if i not in 'a,e,i,o,u,é,ó,ú']
variable =  'relaciónn  requiere,  para obtener  el  valor  de un  cierto  término,  el conocimiento de los dos anteriores.'
non_vowels(variable)
```

### Ejercicio 6
>6. Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9). 

```python
# Estudiante 1
```

```python
# Estudiante 2
number_div = [[divisor for divisor in range(2,10) if number % divisor == 0]
           for number in range(1,1001)]
number_div
```
